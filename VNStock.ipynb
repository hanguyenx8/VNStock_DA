2.2 Lấy dữ liệu lịch sử giá của cổ phiếu doanh nghiệp 

```python
MCK="TCB"
def history_data(MCK):
    data = vns.stock_historical_data(symbol=MCK, 
                            start_date="2010-01-01", 
                            end_date='2023-02-25')
    return data

TCB_df = history_data(MCK)
TCB_df.to_csv("TCB_df_upto_2023_02_06.csv")
TCB_df.tail()
```

```python
MCK="FPT"
FPT_df = history_data(MCK)
FPT_df.to_csv("FPT_df_upto_2023_02_06.csv")
FPT_df.tail()
```
